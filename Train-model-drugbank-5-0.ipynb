# **5 Fold CV**

# 1. Singel model
# (RF, XGB, DT, SVM, KNN, catboost, LR)

```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
import gc

from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB # Thêm thư viện NBC
from sklearn.metrics import (f1_score, precision_score, recall_score, 
                             confusion_matrix, log_loss, accuracy_score,
                             average_precision_score)

import xgboost as xgb
import lightgbm as lgb # Thêm thư viện LightGBM
from cuml.ensemble import RandomForestClassifier as cuRF
from cuml.neighbors import KNeighborsClassifier as cuKNN
from cuml.svm import SVC as cuSVC

warnings.filterwarnings("ignore")

class BenchmarkGPU_5Fold:
    def __init__(self, random_state=42, n_splits=5):
        self.seed = random_state
        self.n_splits = n_splits
        self.plot_data = [] 
        self.table_data = []

    def load_data(self, data_path):
        print(f"\n[1] Đang tải file dữ liệu nguyên vẹn...")
        
        df = pd.read_csv(data_path).fillna(0)
        
        cols_drop = ['u', 'v', 'label'] 
        feats = [c for c in df.columns if c not in cols_drop]
        
        self.X_raw = df[feats].to_numpy(dtype=np.float32)
        self.y = df['label'].to_numpy(dtype=np.int32)

        del df
        gc.collect()
        
        print(f" -> Kích thước dữ liệu: {self.X_raw.shape}")

    def get_models(self):
        models = []
        
        models.append(('Decision Tree', DecisionTreeClassifier(random_state=self.seed, max_depth=10)))
        models.append(('Logistic Regression', LogisticRegression(max_iter=1000, n_jobs=-1, random_state=self.seed)))
        
        # Thêm mô hình NBC
        models.append(('NBC', GaussianNB()))
        
        models.append(('kNN', cuKNN(n_neighbors=10)))
        models.append(('SVM', cuSVC(kernel='rbf', probability=True)))
        models.append(('Random Forest', cuRF(n_estimators=100, max_depth=15, random_state=self.seed)))
        
        models.append(('XGBoost', xgb.XGBClassifier(
            n_estimators=100, eval_metric='logloss', tree_method='hist', 
            device='cuda', random_state=self.seed
        )))
        
        # Thay thế CatBoost bằng LightGBM chạy trên GPU
        models.append(('LightGBM', lgb.LGBMClassifier(
            n_estimators=100, max_depth=15, device='gpu', random_state=self.seed
        )))
        
        return dict(models)

    def plot_live_confusion_matrix(self, name, cm):
        plt.figure(figsize=(5, 4))
        
        cm_sum = np.sum(cm, axis=1, keepdims=True)
        cm_perc = cm / cm_sum.astype(float) * 100
        labels = (np.asarray(["{0}\n({1:.1f}%)".format(int(count), perc)
                              for count, perc in zip(cm.flatten(), cm_perc.flatten())])
                 ).reshape(2, 2)

        sns.heatmap(cm, annot=labels, fmt='', cmap='Blues', cbar=False, 
                    annot_kws={"size": 12, "weight": "bold"})
        
        plt.title(f'{name} | Average CM (5-Fold)', fontsize=12, fontweight='bold')
        plt.xticks([0.5, 1.5], [' (0)', ' (1)'])
        plt.yticks([0.5, 1.5], [' (0)', ' (1)'], rotation=0)

        plt.tight_layout()
        plt.show()

    def run_benchmark(self):
        print(f"\n[2] BẮT ĐẦU HUẤN LUYỆN (5-Fold CV cho từng mô hình)...\n")
        
        model_dict = self.get_models()
        skf = StratifiedKFold(n_splits=self.n_splits, shuffle=True, random_state=self.seed)

        for name, model in model_dict.items():
            print(f"=== Đang xử lý mô hình: {name} ===")
            
            run_metrics = {'F1': [], 'Precision': [], 'Recall': [], 'Accuracy': [], 
                           'Specificity': [], 'AUPR': [], 'Loss': [], 'Time': []}
            cms = []

            for fold, (train_idx, val_idx) in enumerate(skf.split(self.X_raw, self.y), 1):
                start_time = time.time()
                
                # Tách dữ liệu theo Fold
                X_train_fold, X_val_fold = self.X_raw[train_idx], self.X_raw[val_idx]
                y_train_fold, y_val_fold = self.y[train_idx], self.y[val_idx]
                
                # Chuẩn hóa bên trong Fold
                scaler = StandardScaler()
                X_train_scaled = scaler.fit_transform(X_train_fold)
                X_val_scaled = scaler.transform(X_val_fold)

                try:
                    model.fit(X_train_scaled, y_train_fold)
                except Exception as e:
                    print(f" -> Lỗi khi train mô hình {name} ở fold {fold}: {e}")
                    continue

                y_pred_val = model.predict(X_val_scaled)

                y_prob_val = None
                try:
                    if hasattr(model, "predict_proba"):
                        y_prob_val = model.predict_proba(X_val_scaled)
                except: pass

                cm = confusion_matrix(y_val_fold, y_pred_val)
                cms.append(cm)
                tn, fp, fn, tp = cm.ravel()
                
                run_metrics['F1'].append(f1_score(y_val_fold, y_pred_val, zero_division=0))
                run_metrics['Precision'].append(precision_score(y_val_fold, y_pred_val, zero_division=0))
                run_metrics['Recall'].append(recall_score(y_val_fold, y_pred_val, zero_division=0))
                run_metrics['Accuracy'].append(accuracy_score(y_val_fold, y_pred_val))
                run_metrics['Specificity'].append(tn / (tn + fp) if (tn + fp) > 0 else 0)

                val_loss = np.nan
                aupr_val = np.nan
                if y_prob_val is not None:
                    try: 
                        val_loss = log_loss(y_val_fold, y_prob_val)
                        if y_prob_val.shape[1] == 2:
                            aupr_val = average_precision_score(y_val_fold, y_prob_val[:, 1])
                        else:
                            aupr_val = average_precision_score(y_val_fold, y_prob_val)
                    except: pass
                
                run_metrics['Loss'].append(val_loss if not np.isnan(val_loss) else 99.0)
                run_metrics['AUPR'].append(aupr_val if not np.isnan(aupr_val) else 0.0)
                run_time = time.time() - start_time
                run_metrics['Time'].append(run_time)
                
                print(f"    -> Fold {fold} xong! (Time: {run_time:.2f}s | F1: {run_metrics['F1'][-1]:.4f})")
                
                del X_train_fold, X_val_fold, X_train_scaled, X_val_scaled
                gc.collect()

            avg_cm = np.mean(cms, axis=0) 
            self.plot_live_confusion_matrix(name, avg_cm)

            mean_vals = {k: np.mean(v) for k, v in run_metrics.items()}
            self.plot_data.append({
                'Model': name,
                'Loss': mean_vals['Loss'],
                'Accuracy': mean_vals['Accuracy']
            })

            fmt = lambda k: f"{np.mean(run_metrics[k])*100:.2f}"
            loss_fmt = lambda k: f"{np.mean(run_metrics[k]):.4f}" # Giữ nguyên format nhỏ cho Loss
            
            self.table_data.append({
                'Model': name,
                'F1': fmt('F1'),
                'AUPR': fmt('AUPR'),
                'Accuracy': fmt('Accuracy'),
                'Precision': fmt('Precision'),
                'Recall': fmt('Recall'),
                'Specificity': fmt('Specificity'),
                'Loss': loss_fmt('Loss'),
                'Mean_AUPR_Sort': mean_vals['AUPR'] 
            })
        
        print("\n[5] BẢNG TỔNG HỢP CHI TIẾT SAU 5-FOLD CV (Giá trị trung bình)")
        if self.table_data:
            res_df = pd.DataFrame(self.table_data).sort_values(by='Mean_AUPR_Sort', ascending=False)
            res_df_display = res_df.drop(columns=['Mean_AUPR_Sort']) 
            print(res_df_display.to_string(index=False))

    def plot_final_loss_chart(self):
        if not self.plot_data: return
        df_plot = pd.DataFrame(self.plot_data)
        df_plot = df_plot[df_plot['Loss'] < 90].sort_values(by='Loss', ascending=True)

        plt.figure(figsize=(10, 5))
        plt.plot(df_plot['Model'], df_plot['Loss'], marker='o', linestyle='-', color='#d62728', markersize=8, linewidth=2)
        
        plt.ylabel('Mean Log Loss')
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.xticks(rotation=45)
        
        for i in range(len(df_plot)):
            val = df_plot['Loss'].iloc[i]
            plt.annotate(f"{val:.3f}", (df_plot['Model'].iloc[i], val), 
                         textcoords="offset points", xytext=(0,10), ha='center', fontweight='bold')

        plt.tight_layout()
        plt.show()

    def plot_final_accuracy_chart(self):
        if not self.plot_data: return
        df = pd.DataFrame(self.plot_data).sort_values(by='Accuracy', ascending=False)

        plt.figure(figsize=(10, 6))
        plt.plot(df['Model'], df['Accuracy'], marker='o', linestyle='-', color='#1f77b4', markersize=8, linewidth=2)
        
        plt.xlabel('Mô hình', fontsize=12)
        plt.ylabel('Mean Accuracy', fontsize=12)
        plt.ylim(0, 1.1) 
        plt.grid(True, linestyle='--', alpha=0.5)
        plt.xticks(rotation=45)

        for i in range(len(df)):
            val = df['Accuracy'].iloc[i]
            plt.annotate(f'{val:.2%}', (df['Model'].iloc[i], val), 
                         textcoords="offset points", xytext=(0, 10), 
                         ha='center', fontweight='bold', fontsize=10)

        plt.tight_layout()
        plt.show()


if __name__ == "__main__":
    DATA_FILE = "/kaggle/input/datasets/vungocthien/dataset-trainmodel/Dataset_Final.csv" 
    
    if os.path.exists(DATA_FILE):
        bencher = BenchmarkGPU_5Fold(random_state=42)
        bencher.load_data(DATA_FILE)
        bencher.run_benchmark()
    else:
        print("Lỗi: Không tìm thấy file dữ liệu. Hãy kiểm tra lại đường dẫn.")
```

```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
import gc

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (f1_score, precision_score, recall_score, 
                             confusion_matrix, log_loss, accuracy_score,
                             average_precision_score)

import xgboost as xgb
import lightgbm as lgb # Thêm thư viện LightGBM
from cuml.ensemble import RandomForestClassifier as cuRF
from cuml.neighbors import KNeighborsClassifier as cuKNN
from cuml.svm import SVC as cuSVC

warnings.filterwarnings("ignore")

class BenchmarkGPU_RandomSeeds:
    def __init__(self):
        self.table_data = []
        self.seeds = [7, 99, 123, 999, 2024] 

    def load_data(self, train_path, test_path):
        print(f"\n[1] Đang tải và xử lý dữ liệu Train & Test...")
        
        df_train = pd.read_csv(train_path).fillna(0)
        df_test = pd.read_csv(test_path).fillna(0)
        
        cols_drop = ['u', 'v', 'label'] 
        feats = [c for c in df_train.columns if c not in cols_drop]
        
        self.X_train = df_train[feats].to_numpy(dtype=np.float32)
        self.y_train = df_train['label'].to_numpy(dtype=np.int32)
        self.X_test = df_test[feats].to_numpy(dtype=np.float32)
        self.y_test = df_test['label'].to_numpy(dtype=np.int32)

        del df_train, df_test
        gc.collect()

        scaler = StandardScaler()
        self.X_train = scaler.fit_transform(self.X_train)
        self.X_test = scaler.transform(self.X_test)
        
        print(f" -> Kích thước Train: {self.X_train.shape}")
        print(f" -> Kích thước Test:  {self.X_test.shape}")

    def get_models(self, seed):
        models = []
        models.append(('NBC', GaussianNB()))
        models.append(('KNN', cuKNN(n_neighbors=10)))
        models.append(('DT', DecisionTreeClassifier(random_state=seed, max_depth=10)))
        models.append(('SVM', cuSVC(kernel='rbf', probability=True)))
        models.append(('LR', LogisticRegression(max_iter=1000, n_jobs=-1, random_state=seed)))
        models.append(('RF', cuRF(n_estimators=100, max_depth=15, random_state=seed)))

        xgb_model = xgb.XGBClassifier(
            n_estimators=100, eval_metric='logloss',
            device='cuda', tree_method='hist', random_state=seed
        )
        models.append(('XGB', xgb_model))
        
        lgbm_model = lgb.LGBMClassifier(
            n_estimators=100, max_depth=15, device='gpu', random_state=seed
        )
        models.append(('LG', lgbm_model))
        
        return dict(models)

    def plot_live_confusion_matrix(self, name, cm):
        plt.figure(figsize=(5, 4))
        
        cm_sum = np.sum(cm, axis=1, keepdims=True)
        cm_perc = cm / cm_sum.astype(float) * 100
        labels = (np.asarray(["{0}\n({1:.2f}%)".format(int(count), perc)
                              for count, perc in zip(cm.flatten(), cm_perc.flatten())])
                 ).reshape(2, 2)

        sns.heatmap(cm, annot=labels, fmt='', cmap='Blues', cbar=False, 
                    annot_kws={"size": 12, "weight": "bold"})
        
        plt.title(f'{name} | Average CM (5 seeds)', fontsize=12, fontweight='bold')
        
        plt.xticks([0.5, 1.5], [' (0)', ' (1)'])
        plt.yticks([0.5, 1.5], [' (0)', ' (1)'], rotation=0)

        plt.tight_layout()
        plt.show()

    def run_benchmark(self):
        print(f"\n[2] BẮT ĐẦU HUẤN LUYỆN (Chạy lặp 5 lần với các Random Seeds khác nhau)...")
        
        model_names = list(self.get_models(seed=0).keys())

        for name in model_names:
            print(f"=== Đang xử lý mô hình: {name} ===")
            run_metrics = {'F1': [], 'Precision': [], 'Recall': [], 'Accuracy': [], 
                           'Specificity': [], 'AUPR': [], 'Loss': [], 'Time': []}
            cms = []

            for seed in self.seeds:
                start_time = time.time()
                
                model = self.get_models(seed)[name]

                try:
                    model.fit(self.X_train, self.y_train)
                except Exception as e:
                    print(f" -> Lỗi khi train mô hình {name} ở seed {seed}: {e}")
                    continue

                y_pred_test = model.predict(self.X_test)

                y_prob_test = None
                try:
                    if hasattr(model, "predict_proba"):
                        y_prob_test = model.predict_proba(self.X_test)
                except: pass

                cm = confusion_matrix(self.y_test, y_pred_test)
                cms.append(cm)
                tn, fp, fn, tp = cm.ravel()
                
                run_metrics['F1'].append(f1_score(self.y_test, y_pred_test, zero_division=0))
                run_metrics['Precision'].append(precision_score(self.y_test, y_pred_test, zero_division=0))
                run_metrics['Recall'].append(recall_score(self.y_test, y_pred_test, zero_division=0))
                run_metrics['Accuracy'].append(accuracy_score(self.y_test, y_pred_test))
                run_metrics['Specificity'].append(tn / (tn + fp) if (tn + fp) > 0 else 0)

                test_loss = np.nan
                aupr_test = np.nan
                if y_prob_test is not None:
                    try: 
                        test_loss = log_loss(self.y_test, y_prob_test)
                        if y_prob_test.shape[1] == 2:
                            aupr_test = average_precision_score(self.y_test, y_prob_test[:, 1])
                        else:
                            aupr_test = average_precision_score(self.y_test, y_prob_test)
                    except: pass
                
                run_metrics['Loss'].append(test_loss if not np.isnan(test_loss) else 99.0)
                run_metrics['AUPR'].append(aupr_test if not np.isnan(aupr_test) else 0.0)
                run_metrics['Time'].append(time.time() - start_time)
                
                gc.collect()

            avg_cm = np.mean(cms, axis=0) 
            self.plot_live_confusion_matrix(name, avg_cm)

            mean_vals = {k: np.mean(v) for k, v in run_metrics.items()}
            
            fmt = lambda k: f"{mean_vals[k]*100:.2f}"
            loss_fmt = lambda k: f"{mean_vals[k]:.4f}" 
            
            self.table_data.append({
                'Model': name,
                'F1': fmt('F1'),
                'AUPR': fmt('AUPR'),
                'Accuracy': fmt('Accuracy'),
                'Precision': fmt('Precision'),
                'Recall': fmt('Recall'),
                'Specificity': fmt('Specificity'),
                'Loss': loss_fmt('Loss'),
                'Mean_AUPR_Sort': mean_vals['AUPR'] 
            })

        print("\n[3] BẢNG TỔNG HỢP CHI TIẾT (Giá trị trung bình sau 5 lần chạy Seed)")
        if self.table_data:
            res_df = pd.DataFrame(self.table_data).sort_values(by='Mean_AUPR_Sort', ascending=False)
            res_df_display = res_df.drop(columns=['Mean_AUPR_Sort']) 
            print(res_df_display.to_string(index=False))

if __name__ == "__main__":
    TRAIN_FILE = "/kaggle/input/datasets/vungocthien/dataset-trainmodel/train_data.csv" 
    TEST_FILE = "/kaggle/input/datasets/vungocthien/dataset-trainmodel/test_data.csv"   
    
    if os.path.exists(TRAIN_FILE) and os.path.exists(TEST_FILE):
        bencher = BenchmarkGPU_RandomSeeds()
        bencher.load_data(TRAIN_FILE, TEST_FILE)
        bencher.run_benchmark()
    else:
        print("Lỗi: Không tìm thấy file dữ liệu Train hoặc Test. Hãy kiểm tra lại đường dẫn.")
```

```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
import gc

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (f1_score, precision_score, recall_score, 
                             confusion_matrix, log_loss, accuracy_score,
                             average_precision_score)

import xgboost as xgb
import lightgbm as lgb # Thêm thư viện LightGBM
from cuml.ensemble import RandomForestClassifier as cuRF
from cuml.neighbors import KNeighborsClassifier as cuKNN
from cuml.svm import SVC as cuSVC

warnings.filterwarnings("ignore")

class BenchmarkGPU_RandomSeeds:
    def __init__(self):
        self.table_data = []
        self.seeds = [7, 99, 123, 999, 2024] 

    def load_data(self, train_path, test_path):
        print(f"\n[1] Đang tải và xử lý dữ liệu Train & Test...")
        
        df_train = pd.read_csv(train_path).fillna(0)
        df_test = pd.read_csv(test_path).fillna(0)
        
        cols_drop = ['u', 'v', 'label'] 
        feats = [c for c in df_train.columns if c not in cols_drop]
        
        self.X_train = df_train[feats].to_numpy(dtype=np.float32)
        self.y_train = df_train['label'].to_numpy(dtype=np.int32)
        self.X_test = df_test[feats].to_numpy(dtype=np.float32)
        self.y_test = df_test['label'].to_numpy(dtype=np.int32)

        del df_train, df_test
        gc.collect()

        scaler = StandardScaler()
        self.X_train = scaler.fit_transform(self.X_train)
        self.X_test = scaler.transform(self.X_test)
        
        print(f" -> Kích thước Train: {self.X_train.shape}")
        print(f" -> Kích thước Test:  {self.X_test.shape}")

    def get_models(self, seed):
        models = []
        models.append(('NBC', GaussianNB()))
        models.append(('KNN', cuKNN(n_neighbors=10)))
        models.append(('DT', DecisionTreeClassifier(random_state=seed, max_depth=10)))
        models.append(('SVM', cuSVC(kernel='rbf', probability=True)))
        models.append(('LR', LogisticRegression(max_iter=1000, n_jobs=-1, random_state=seed)))
        models.append(('RF', cuRF(n_estimators=100, max_depth=15, random_state=seed)))

        xgb_model = xgb.XGBClassifier(
            n_estimators=100, eval_metric='logloss',
            device='cuda', tree_method='hist', random_state=seed
        )
        models.append(('XGB', xgb_model))
        
        lgbm_model = lgb.LGBMClassifier(
            n_estimators=100, max_depth=15, device='gpu', random_state=seed
        )
        models.append(('LG', lgbm_model))
        
        return dict(models)

    def plot_live_confusion_matrix(self, name, cm):
        plt.figure(figsize=(5, 4))
        
        cm_sum = np.sum(cm, axis=1, keepdims=True)
        cm_perc = cm / cm_sum.astype(float) * 100
        labels = (np.asarray(["{0}\n({1:.2f}%)".format(int(count), perc)
                              for count, perc in zip(cm.flatten(), cm_perc.flatten())])
                 ).reshape(2, 2)

        sns.heatmap(cm, annot=labels, fmt='', cmap='Blues', cbar=False, 
                    annot_kws={"size": 12, "weight": "bold"})
        
        plt.title(f'{name} | Average CM (5 seeds)', fontsize=12, fontweight='bold')
        
        plt.xticks([0.5, 1.5], [' (0)', ' (1)'])
        plt.yticks([0.5, 1.5], [' (0)', ' (1)'], rotation=0)

        plt.tight_layout()
        plt.show()

    def run_benchmark(self):
        print(f"\n[2] BẮT ĐẦU HUẤN LUYỆN (Chạy lặp 5 lần với các Random Seeds khác nhau)...")
        
        model_names = list(self.get_models(seed=0).keys())

        for name in model_names:
            print(f"=== Đang xử lý mô hình: {name} ===")
            run_metrics = {'F1': [], 'Precision': [], 'Recall': [], 'Accuracy': [], 
                           'Specificity': [], 'AUPR': [], 'Loss': [], 'Time': []}
            cms = []

            for seed in self.seeds:
                start_time = time.time()
                
                model = self.get_models(seed)[name]

                try:
                    model.fit(self.X_train, self.y_train)
                except Exception as e:
                    print(f" -> Lỗi khi train mô hình {name} ở seed {seed}: {e}")
                    continue

                y_pred_test = model.predict(self.X_test)

                y_prob_test = None
                try:
                    if hasattr(model, "predict_proba"):
                        y_prob_test = model.predict_proba(self.X_test)
                except: pass

                cm = confusion_matrix(self.y_test, y_pred_test)
                cms.append(cm)
                tn, fp, fn, tp = cm.ravel()
                
                run_metrics['F1'].append(f1_score(self.y_test, y_pred_test, zero_division=0))
                run_metrics['Precision'].append(precision_score(self.y_test, y_pred_test, zero_division=0))
                run_metrics['Recall'].append(recall_score(self.y_test, y_pred_test, zero_division=0))
                run_metrics['Accuracy'].append(accuracy_score(self.y_test, y_pred_test))
                run_metrics['Specificity'].append(tn / (tn + fp) if (tn + fp) > 0 else 0)

                test_loss = np.nan
                aupr_test = np.nan
                if y_prob_test is not None:
                    try: 
                        test_loss = log_loss(self.y_test, y_prob_test)
                        if y_prob_test.shape[1] == 2:
                            aupr_test = average_precision_score(self.y_test, y_prob_test[:, 1])
                        else:
                            aupr_test = average_precision_score(self.y_test, y_prob_test)
                    except: pass
                
                run_metrics['Loss'].append(test_loss if not np.isnan(test_loss) else 99.0)
                run_metrics['AUPR'].append(aupr_test if not np.isnan(aupr_test) else 0.0)
                run_metrics['Time'].append(time.time() - start_time)
                
                gc.collect()

            avg_cm = np.mean(cms, axis=0) 
            self.plot_live_confusion_matrix(name, avg_cm)

            mean_vals = {k: np.mean(v) for k, v in run_metrics.items()}
            
            fmt = lambda k: f"{mean_vals[k]*100:.2f}"
            loss_fmt = lambda k: f"{mean_vals[k]:.4f}" 
            
            self.table_data.append({
                'Model': name,
                'F1': fmt('F1'),
                'AUPR': fmt('AUPR'),
                'Accuracy': fmt('Accuracy'),
                'Precision': fmt('Precision'),
                'Recall': fmt('Recall'),
                'Specificity': fmt('Specificity'),
                'Loss': loss_fmt('Loss'),
                'Mean_AUPR_Sort': mean_vals['AUPR'] 
            })

        print("\n[3] BẢNG TỔNG HỢP CHI TIẾT (Giá trị trung bình sau 5 lần chạy Seed)")
        if self.table_data:
            res_df = pd.DataFrame(self.table_data).sort_values(by='Mean_AUPR_Sort', ascending=False)
            res_df_display = res_df.drop(columns=['Mean_AUPR_Sort']) 
            print(res_df_display.to_string(index=False))

if __name__ == "__main__":
    TRAIN_FILE = "/kaggle/working/9FT_train.csv" 
    TEST_FILE = "/kaggle/working/9FT_test.csv"   
    
    if os.path.exists(TRAIN_FILE) and os.path.exists(TEST_FILE):
        bencher = BenchmarkGPU_RandomSeeds()
        bencher.load_data(TRAIN_FILE, TEST_FILE)
        bencher.run_benchmark()
    else:
        print("Lỗi: Không tìm thấy file dữ liệu Train hoặc Test. Hãy kiểm tra lại đường dẫn.")
```

```python
import pandas as pd
df = pd.read_csv('/kaggle/input/datasets/vungocthien/dataset-trainmodel/Dataset_Final.csv')
df = df.drop(["sim_ade", "sim_atc"], axis=1)
df.to_csv("10FT.csv", index=False)
df
```

```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
import gc

from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (f1_score, precision_score, recall_score, 
                             confusion_matrix, log_loss, accuracy_score,
                             average_precision_score)

import xgboost as xgb
import lightgbm as lgb 

warnings.filterwarnings("ignore")

class BenchmarkCPU_5FoldCV:
    def __init__(self, random_state=42, n_splits=5, max_samples=20000):
        self.table_data = []
        self.seed = random_state
        self.n_splits = n_splits
        self.max_samples = max_samples # Giới hạn số dòng cho các mô hình chậm

    def load_data(self, data_path):
        print(f"\n[1] Đang tải file dữ liệu nguyên vẹn để chạy 5-Fold CV...")
        
        df = pd.read_csv(data_path).fillna(0)
        
        cols_drop = ['u', 'v', 'label'] 
        feats = [c for c in df.columns if c not in cols_drop]
        
        # CPU xử lý mượt nhất với kiểu float64 mặc định
        self.X_raw = df[feats].to_numpy()
        self.y = df['label'].to_numpy(dtype=np.int32)

        del df
        gc.collect()
        
        print(f" -> Kích thước toàn bộ dữ liệu: {self.X_raw.shape}")

    def get_models(self):
        models = []
        models.append(('NBC', GaussianNB()))
        models.append(('KNN', KNeighborsClassifier(n_neighbors=10, n_jobs=-1)))
        models.append(('DT', DecisionTreeClassifier(random_state=self.seed, max_depth=10)))
        models.append(('LR', LogisticRegression(max_iter=1000, n_jobs=-1, random_state=self.seed)))
        models.append(('RF', RandomForestClassifier(n_estimators=100, max_depth=15, n_jobs=-1, random_state=self.seed)))

        xgb_model = xgb.XGBClassifier(
            n_estimators=100, eval_metric='logloss',
            tree_method='hist', n_jobs=-1, random_state=self.seed
        )
        models.append(('XGB', xgb_model))
        
        lgbm_model = lgb.LGBMClassifier(
            n_estimators=100, max_depth=15, n_jobs=-1, random_state=self.seed
        )
        models.append(('LG', lgbm_model))
        
        # SVM trên CPU
        models.append(('SVM', SVC(kernel='rbf', probability=True, random_state=self.seed)))
        
        return dict(models)

    def plot_live_confusion_matrix(self, name, cm):
        plt.figure(figsize=(5, 4))
        
        cm_sum = np.sum(cm, axis=1, keepdims=True)
        cm_perc = cm / cm_sum.astype(float) * 100
        labels = (np.asarray(["{0}\n({1:.2f}%)".format(int(count), perc)
                              for count, perc in zip(cm.flatten(), cm_perc.flatten())])
                 ).reshape(2, 2)

        sns.heatmap(cm, annot=labels, fmt='', cmap='Blues', cbar=False, 
                    annot_kws={"size": 12, "weight": "bold"})
        
        plt.title(f'{name} | Average CM (5 Folds)', fontsize=12, fontweight='bold')
        plt.xticks([0.5, 1.5], [' (0)', ' (1)'])
        plt.yticks([0.5, 1.5], [' (0)', ' (1)'], rotation=0)

        plt.tight_layout()
        plt.show()

    def run_benchmark(self):
        print(f"\n[2] BẮT ĐẦU HUẤN LUYỆN TRÊN CPU (5-Fold CV)...")
        
        model_dict = self.get_models()
        skf = StratifiedKFold(n_splits=self.n_splits, shuffle=True, random_state=self.seed)

        for name, model in model_dict.items():
            print(f"\n=== Đang xử lý mô hình: {name} ===")
            
            run_metrics = {'F1': [], 'Precision': [], 'Recall': [], 'Accuracy': [], 
                           'Specificity': [], 'AUPR': [], 'Loss': [], 'Time': []}
            cms = []

            for fold, (train_idx, val_idx) in enumerate(skf.split(self.X_raw, self.y), 1):
                start_time = time.time()
                
                # Tách dữ liệu theo Fold
                X_train_fold, X_val_fold = self.X_raw[train_idx], self.X_raw[val_idx]
                y_train_fold, y_val_fold = self.y[train_idx], self.y[val_idx]
                
                # Chuẩn hóa bên trong Fold để chống rò rỉ dữ liệu (Leakage)
                scaler = StandardScaler()
                X_train_scaled = scaler.fit_transform(X_train_fold)
                X_val_scaled = scaler.transform(X_val_fold)
                
                # --- CƠ CHẾ CẮT GIẢM DỮ LIỆU CHO SVM VÀ KNN ---
                X_train_fit, y_train_fit = X_train_scaled, y_train_fold
                
                if name in ['SVM'] and len(X_train_scaled) > self.max_samples:
                    # Dùng seed + fold để mỗi fold lấy ngẫu nhiên một mảng dữ liệu khác nhau
                    np.random.seed(self.seed + fold) 
                    idx = np.random.choice(len(X_train_scaled), self.max_samples, replace=False)
                    X_train_fit = X_train_scaled[idx]
                    y_train_fit = y_train_fold[idx]
                    
                try:
                    # Ép học trên tập dữ liệu đã cắt giảm (hoặc tập gốc nếu là các mô hình khác)
                    model.fit(X_train_fit, y_train_fit)
                except Exception as e:
                    print(f" -> Lỗi khi train mô hình {name} ở fold {fold}: {e}")
                    continue

                y_pred_val = model.predict(X_val_scaled)

                y_prob_val = None
                try:
                    if hasattr(model, "predict_proba"):
                        y_prob_val = model.predict_proba(X_val_scaled)
                except: pass

                cm = confusion_matrix(y_val_fold, y_pred_val)
                cms.append(cm)
                tn, fp, fn, tp = cm.ravel()
                
                run_metrics['F1'].append(f1_score(y_val_fold, y_pred_val, zero_division=0))
                run_metrics['Precision'].append(precision_score(y_val_fold, y_pred_val, zero_division=0))
                run_metrics['Recall'].append(recall_score(y_val_fold, y_pred_val, zero_division=0))
                run_metrics['Accuracy'].append(accuracy_score(y_val_fold, y_pred_val))
                run_metrics['Specificity'].append(tn / (tn + fp) if (tn + fp) > 0 else 0)

                val_loss = np.nan
                aupr_val = np.nan
                if y_prob_val is not None:
                    try: 
                        val_loss = log_loss(y_val_fold, y_prob_val)
                        if y_prob_val.shape[1] == 2:
                            aupr_val = average_precision_score(y_val_fold, y_prob_val[:, 1])
                        else:
                            aupr_val = average_precision_score(y_val_fold, y_prob_val)
                    except: pass
                
                run_metrics['Loss'].append(val_loss if not np.isnan(val_loss) else 99.0)
                run_metrics['AUPR'].append(aupr_val if not np.isnan(aupr_val) else 0.0)
                
                run_time = time.time() - start_time
                run_metrics['Time'].append(run_time)
                
                print(f"    -> Fold {fold} xong! (Time: {run_time:.2f}s | Acc: {run_metrics['Accuracy'][-1]*100:.2f}%)")
                
                del X_train_fold, X_val_fold, X_train_scaled, X_val_scaled
                gc.collect()

            avg_cm = np.mean(cms, axis=0) 
            self.plot_live_confusion_matrix(name, avg_cm)

            mean_vals = {k: np.mean(v) for k, v in run_metrics.items()}
            
            fmt = lambda k: f"{mean_vals[k]*100:.2f}"
            loss_fmt = lambda k: f"{mean_vals[k]:.4f}" 
            
            self.table_data.append({
                'Model': name,
                'F1': fmt('F1'),
                'AUPR': fmt('AUPR'),
                'Accuracy': fmt('Accuracy'),
                'Precision': fmt('Precision'),
                'Recall': fmt('Recall'),
                'Specificity': fmt('Specificity'),
                'Loss': loss_fmt('Loss'),
                'Mean_AUPR_Sort': mean_vals['AUPR'] 
            })

        print("\n[3] BẢNG TỔNG HỢP CHI TIẾT (Giá trị trung bình 5-Fold CV)")
        if self.table_data:
            res_df = pd.DataFrame(self.table_data).sort_values(by='Mean_AUPR_Sort', ascending=False)
            res_df_display = res_df.drop(columns=['Mean_AUPR_Sort']) 
            print(res_df_display.to_string(index=False))

if __name__ == "__main__":
    DATA_FILE = "/kaggle/working/10FT.csv" 
    
    if os.path.exists(DATA_FILE):
        
        bencher = BenchmarkCPU_5FoldCV(random_state=42, n_splits=5, max_samples=30000)
        bencher.load_data(DATA_FILE)
        bencher.run_benchmark()
    else:
        print(f"Lỗi: Không tìm thấy file dữ liệu tại đường dẫn: {DATA_FILE}")
```
